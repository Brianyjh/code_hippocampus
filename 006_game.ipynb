```python
import pandas as pd
```

```python
order_products = pd.read_csv("./instacart/order_products__prior.csv")
products = pd.read_csv("./instacart/products.csv")
orders = pd.read_csv("./instacart/orders.csv")
aisles = pd.read_csv("./instacart/aisles.csv")
```

```python
table1 = pd.merge(aisles,products,on="aisle_id")
table2 = pd.merge(aisles,products,on=["aisle_id","aisle_id"])
```

```python
table3 = pd.merge(table1,order_products,on="product_id")
```

```python
table3
```

```python
table4 = pd.merge(table3,orders,on="order_id")
```

```python
table4
```

```python
# 交叉表
table_c = pd.crosstab(table4["aisle"],table4["user_id"])
```

```python
table_T = table_c.T
table_T
```

```python
table = table_T[:10000]
table
```
